## Contains orgs (opposite of above)

```python
# Filter the DataFrame to get the opposite of no_orgs
# and check for semicolon in 'responses' or 'article_id' in 'no_orgs'
contains_orgs = df[~(df["article_id"].isin(no_orgs["article_id"]))]
contains_orgs = contains_orgs[~(contains_orgs["responses"].str.contains(r"org:[^;\n]*;"))]

# Print the length and head of the contains_orgs DataFrame
print("Length:", len(contains_orgs))
contains_orgs.head()
```

### Distribution of unique items
#### unique orgs

```python
# Convert string representations to dictionaries
contains_orgs['responses_PP_ast'] = contains_orgs['responses_PP'].apply(ast.literal_eval)

# Extract 'org' values and count the unique organizations
org_counts = contains_orgs['responses_PP_ast'].apply(lambda x: x.get('org', np.nan)).dropna().value_counts()

# Create a DataFrame with 'ORG' and 'count' columns, excluding blank org values
org_dist = pd.DataFrame({'ORG': org_counts.index, 'count': org_counts.values})

# Remove leading and trailing whitespace from 'ORG' column
org_dist['ORG'] = org_dist['ORG'].str.strip()

# Filter out rows with empty 'ORG' values
org_dist = org_dist[org_dist['ORG'] != '']

# Write the DataFrame to an Excel file
org_dist.to_excel('./results/aylien_responses_summary/org_dist.xlsx', index=False)

print(len(org_dist))
org_dist.head()
```

### all unique items

```python
# Function to split values and count unique items separately
def count_unique_items(field):
    unique_items = contains_orgs['responses_PP_ast'].apply(lambda x: str(x.get(field)))
    unique_items = unique_items.dropna().str.split(';').explode().str.strip()
    unique_items_count = unique_items.nunique()
    return unique_items_count

# Fields for which you want to get the count of unique items
fields = ['org', 'country', 'state', 'city', 'industry', 'risks', 'items_sold', 'service_provided', 'business_relations', 'article_id', 'source']

# List to store item and count pairs
summary_list = []

# Iterate over the fields and get the count of unique items
for field in fields:
    unique_item_count = count_unique_items(field)
    summary_list.append((field, unique_item_count))

# Create a DataFrame from the summary list
summary_df = pd.DataFrame(summary_list, columns=['item', 'count'])

# Return the summary DataFrame
summary_df
```

## Longest responses

```python
# Calculate the length of data in the 'responses' column
df['response_length'] = df['responses'].str.len()

# Get the top 10 longest responses
top_10_longest_responses = df.nlargest(10, 'response_length')

# Print the top 10 longest responses
print(len(top_10_longest_responses))
top_10_longest_responses.head()

# export
top_10_longest_responses.to_excel('./results/aylien_responses_summary/top_10_longest_responses.xlsx')
```

### Longset articles

```python
# Calculate the length of data in the 'body' column
df['body_length'] = df['body'].str.len()

# Get the top 10 longest values in the 'body' column
top_10_longest_articles = df.nlargest(10, 'body_length')

# Print the top 10 longest responses
print(len(top_10_longest_articles))
top_10_longest_articles

# export
top_10_longest_articles.to_excel('./results/aylien_responses_summary/top_10_longest_articles.xlsx')
```

### Shortest articles

```python
# Calculate the length of data in the 'body' column
df['body_length'] = df['body'].str.len()

# Get the indices of the top 10 shortest articles in ascending order
indices = df.nsmallest(10, 'body_length').index

# Retrieve the top 10 shortest articles using the sorted indices
top_10_shortest_articles = df.loc[indices]
top_10_shortest_articles

# export
top_10_shortest_articles.to_excel('./results/aylien_responses_summary/top_10_shortest_articles.xlsx')
```
